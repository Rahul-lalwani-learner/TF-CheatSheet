```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import random
from tensorflow.keras.datasets import boston_housing
import wandb
```

```python
wandb.login()
```

```python
(X_train , y_train), (X_test, y_test) = boston_housing.load_data()
len(X_train), len(y_train), len(X_test), len(y_test)
```

```python
X_train[0].shape
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.losses import mae, mse
from tensorflow.keras.optimizers import Adam, SGD
from wandb.keras import WandbMetricsLogger, WandbModelCheckpoint
```

## Building first model

```python
tf.random.set_seed(42)

wandb.init(
    project="Model-1-boston-data",
    # (optional) set entity to specify your username or team name
    # entity="my_team",
    config={
        'layer_1': 32,
        'activation_1': 'relu',
        'layer_2': 16,
        'activation_2': 'linear',
        'layer_3': 1,
        'loss': mae,
        'optimizer_1': Adam(),
        'optimizer_2': SGD(),
        'learning_rate': 0.01,
        'metric': 'mse',
        'epochs': 10
    },
)
config = wandb.config
model_1 = Sequential([
    Dense(config.layer_1, input_shape = [13], activation = config.activation_1),
    Dense(config.layer_1, activation = config.activation_1),
    Dense(config.layer_2, activation = config.activation_1),
    Dense(config.layer_3, activation = config.activation_2)
])

wandb_callback = [
    WandbMetricsLogger(),
    WandbModelCheckpoint(filepath="my_model_{epoch:02d}"),
]
model_1.compile(loss = mae,
                optimizer = Adam(),
                metrics = ['mse'])
model_1.fit(X_train,y_train,
            epochs = 10,
            callbacks = wandb_callback)

wandb.finish()
```

```python
model_1.evaluate(X_test,y_test)
```

```python
model_1.predict(X_test[:10])
```

### let's use SGD optimizer this time

```python
wandb.init(
    project = "Model-1-boston-data"
)

wandb_callback =[
    WandbMetricsLogger(),
    WandbModelCheckpoint(filepath = "model_1/epoch_{epoch:02d}")
]

model_2 = Sequential([
    Dense(32, activation = 'relu'),
    Dense(32, activation = 'relu'),
    Dense(16, activation = 'relu'),
    Dense(1, activation = 'linear')
])
model_2.compile(loss = mae,
                optimizer = SGD(learning_rate = 0.01),
                metrics = ['mse'])
model_2.fit(X_train,
            y_train,
            epochs =10,
            callbacks = wandb_callback
            )
```

```python
model_2.evaluate(X_test, y_test)
```

## model_1 performed better so why not train it for longer uisng Adam optimizer

```python
model_3 = Sequential([
    Dense(32, activation = 'relu'),
    Dense(32, activation = 'relu'),
    Dense(16, activation = 'relu'),
    Dense(1, activation = 'linear')
])

model_3.compile(loss = mae,
                optimizer = Adam(),
                metrics = ['mse'])

wandb.init(project = "Model-1-boston-data")

EarlyStopping = tf.keras.callbacks.EarlyStopping(monitor = 'val_loss', patience = 3)
model_3.fit(X_train , y_train,
            epochs = 50,
            validation_data = (X_test, y_test),
            callbacks = [
                WandbMetricsLogger(),
                WandbModelCheckpoint(filepath = "/content/model_3/epochs_{epoch:02d}")
            ])
wandb.finish()
```

```python
model_3.evaluate(X_test, y_test)
```

## Lastly let's change the learning rate to 0.01 and see how it effects

```python
tf.random.set_seed(42)

model_4 = Sequential([
    Dense(32, activation = 'relu'),
    Dense(32, activation = 'relu'),
    Dense(16, activation = 'relu'),
    Dense(1, activation = 'linear')
])

model_4.compile(loss = mae,
                optimizer = Adam(learning_rate = 0.01),
                metrics = ['mse'])

wandb.init(project = "Model-1-boston-data", name = "Learning_rate_0.01")

model_4.fit(X_train , y_train,
            epochs = 50,
            validation_data = (X_test, y_test),
            callbacks = [
                WandbMetricsLogger(),
                WandbModelCheckpoint(filepath = "content/model_4/epochs_{epoch:02d}")
            ])
wandb.finish()
```
